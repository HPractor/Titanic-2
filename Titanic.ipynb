# **Splitting the dataset into training and testing sets**

```python
train,test=train_test_split(a,test_size=0.3)
```

# **Performing Training and Testing**

```python
train_x=train.drop(['Survived'],axis=1)
```

```python
train_y=train['Survived']
```

```python
test_x=test.drop(['Survived'],axis=1)
```

```python
test_y=test['Survived']
```

# **Linear Regression**

```python
model = LinearRegression()
model.fit(train_x, train_y)
```

```python
prediction = model.predict(test_x) # Use the imputed test data
```

```python
lin_reg = LinearRegression()
lin_reg.fit(train_x, train_y)
y_pred_lin = lin_reg.predict(test_x)
print(f'Linear Regression MSE: {mean_squared_error(test_y, y_pred_lin)}')
```

```python
r2_model_score = model.score(test_x,test_y)
print(r2_model_score)
```

```python
# Create a DataFrame to hold actual and predicted values
results = pandas.DataFrame({'Actual': test_y, 'Predicted': prediction})
```

```python
# Create a DataFrame to hold actual and predicted values
results = pandas.DataFrame({'Actual': test_y, 'Predicted': prediction})

# Plotting the graph with actual vs predicted values as coordinates
plt.figure(figsize=(10, 6))
plt.scatter(results['Actual'], results['Predicted'], color='blue', alpha=0.5)

# Plot a 45-degree line to indicate perfect predictions
x = [0, 1]
y = [0, 1]
plt.plot(x, y, color='red', linestyle='--', label='Perfect Prediction')

# Add labels and title
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs Predicted Values (Titanic Dataset)')
plt.legend()

# Display the plot
plt.show()
```
